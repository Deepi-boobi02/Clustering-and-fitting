```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from scipy.stats import skew, kurtosis

# Load the dataset
df = pd.read_csv("netflix_titles.csv")

# Data Preprocessing
df.dropna(subset=['release_year', 'rating'], inplace=True)
df['release_year'] = df['release_year'].astype(int)

# Statistical Moments
mean_year = df['release_year'].mean()
var_year = df['release_year'].var()
skew_year = skew(df['release_year'])
kurt_year = kurtosis(df['release_year'])

print(f"Mean: {mean_year}, Variance: {var_year}, Skewness: {skew_year}, Kurtosis: {kurt_year}")

# Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['release_year'], y=df.index, alpha=0.5)
plt.title("Release Year Distribution (Scatter Plot)")
plt.xlabel("Release Year")
plt.ylabel("Index")
plt.show()

# Histogram Plot
plt.figure(figsize=(8, 5))
sns.histplot(df['release_year'], bins=20, kde=True)
plt.title("Release Year Histogram")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.show()

# Violin Plot
plt.figure(figsize=(8, 5))
sns.violinplot(x=df['release_year'])
plt.title("Release Year Violin Plot")
plt.xlabel("Release Year")
plt.show()


# Clustering
X = df[['release_year']]
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['release_year'], y=df.index, hue=df['cluster'], palette='coolwarm', alpha=0.6)
plt.title("K-Means Clustering on Release Year")
plt.xlabel("Release Year")
plt.ylabel("Index")
plt.show() 
```
